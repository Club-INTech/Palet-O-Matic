## Code filtrage entropy + filtre background/foreground

```python
from skimage import io, filters
#On lit l'image 
image_cercles = io.imread("./image_palets_2.jpg")
from skimage.filters import rank 
import numpy as np
from skimage.color import rgb2gray
#On transforme l'image en niveaux de gris 
image_gray = rgb2gray(image_cercles)
noisy = rank.entropy(image_gray, np.ones((9,9)))
#On fixe un seuil au bruit 
textured_regions_1 = noisy > 4
image = image_cercles.copy()
image[textured_regions_1] = 0

#On trouve la valeur aprs filtrage 
val = filters.threshold_otsu(image_cercles)
image_2 = image.copy()
mask_1 = image_2 > val
mask_2 = image_2 < val
#On met  blanc les pixels respectant le mask 
image_2[mask_1] = 255
#On met  noir les pixels respectant le mask 
image_2[mask_2] = 0
io.imshow(image_2)

io.imsave("image_apres_traitement.png",image_2)
```
